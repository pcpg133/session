# <과제2>
- 구글링도 과제 수행에 포함되어 있습니다. 세션에 소개되지 않은 기능이 필요하다면, 구글링을 통해 구현해주시길 바랍니다.
- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
data2 = pd.read_csv('/content/gdrive/MyDrive/deliver.csv')
data2.head()
```

```python
def month(_x):
  if 3 <= int(_x[5:7]) <= 5:
    return '봄'
  elif 6 <= int(_x[5:7]) <= 8:
    return '여름'
  elif 9 <= int(_x[5:7]) <= 11:
    return '가을'
  else:
    return '겨울'
```

```python
data2['월'] = data2['날짜'].apply(lambda x: x[5:7])
data2.head(30)
```

```python
data2['계절'] = data2['날짜'].apply(month)
data2.head()
```

```python
data2['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
data2['미세먼지(㎍/㎥)'] = data2['미세먼지(㎍/㎥)'].fillna(data2['미세먼지(㎍/㎥)'].mean())
data2.head(30)
```
