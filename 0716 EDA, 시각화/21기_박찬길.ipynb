### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
df.info()
```

```python
missingValue = df.isnull().sum()
missingValue = missingValue[missingValue > 0]
df_miss = pd.DataFrame(missingValue, columns=['missing_value'])
df_miss = df_miss.sort_values(by = 'missing_value', ascending = False)
df_miss['missing_%'] = round(df_miss['missing_value'] / len(df) * 100, 1)
df_miss
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
```

```python
plt.figure(figsize = (15, 4))
sns.barplot(x = df_miss.index, y = 'missing_%', data = df_miss)
plt.xticks(rotation = 45)
plt.show()
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다.
